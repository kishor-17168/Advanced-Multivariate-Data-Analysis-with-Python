# Problem-6:Multidimensional Scaling (MDS) with Iris dataset using python

```python
# Multidimensional Scaling (MDS) Example with Iris dataset
from sklearn.datasets import load_iris
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load default dataset (Iris)
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# Apply MDS (2D projection)
mds = MDS(n_components=2, random_state=42, dissimilarity='euclidean')
X_mds = mds.fit_transform(X)
```

```python
# Create a DataFrame for easy plotting
df_mds = pd.DataFrame(X_mds, columns=['MDS1', 'MDS2'])
df_mds['species'] = [target_names[i] for i in y]
```

```python
# Plot the MDS result
plt.figure(figsize=(8, 6))
for species in target_names:
    subset = df_mds[df_mds['species'] == species]
    plt.scatter(subset['MDS1'], subset['MDS2'], label=species)

plt.title('Multidimensional Scaling (MDS) - Iris Dataset')
plt.xlabel('MDS Dimension 1')
plt.ylabel('MDS Dimension 2')
plt.legend()
plt.show()
```

This scatter plot shows the Iris dataset visualized using Multidimensional Scaling (MDS).

MDS is a technique that takes the distances between data points in a high-dimensional space and represents them in a lower-dimensional space (in this case, two dimensions) while preserving the original distances as much as possible.

* The plot effectively separates the "setosa" species (blue dots) from the other two species.

* The "versicolor" (orange dots) and "virginica" (green dots) species are grouped closer together, but are still somewhat distinguishable.

* The axes "MDS Dimension 1" and "MDS Dimension 2" represent the new coordinates that best capture the relationships between the original data points.
