# Problem-5: DBSCAN using python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Generate synthetic data
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)
```

```python
# Plotting the results
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='rainbow', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

DBSCAN is a density-based clustering algorithm. It groups together points that are closely packed together, marking as outliers points that lie alone in low-density regions.

In this plot:

* The red and purple colors represent two distinct clusters that the DBSCAN algorithm successfully identified.

* The algorithm was able to find these non-linear, crescent-shaped clusters because it's not restricted to finding spherical clusters like K-Means.

* The plot demonstrates DBSCAN's ability to handle complex data shapes and is a good visualization for a report on clustering algorithms.
